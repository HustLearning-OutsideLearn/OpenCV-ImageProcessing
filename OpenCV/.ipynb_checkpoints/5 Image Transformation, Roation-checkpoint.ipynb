```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

resource_path = 'D:\GitCloneProject\OpenCV-ImageProcessing\Resources'
photo_path = resource_path + '\Photos'
video_path = resource_path + '\Videos'
face_path = resource_path + '\Faces'
```

```python
img = cv.imread(photo_path + '\cats 2.jpg')
plt.imshow(img)
```

```python
# Translation
def translate(img, x, y):
    transMat = np.float32([[1,0,x], [0,1,y]])
    dimensions = (img.shape[1], img.shape[0])
    return cv.warpAffine(img, transMat, dimensions)

# -x --> Left
# -y --> Up
# - x --> Right
# -y --> Down

trans = translate(img, 100, 100)
plt.imshow(trans)
```

```python
trans = translate(img, -200, 100)
plt.imshow(trans)
```

```python
# Rotation
def rotate(im, angle, rotPoint=None):
    (height, width) = img.shape[:2]
    
    if rotPoint is None:
        rotPoint = (width//2, height//2)
    rotMat = cv.getRotationMatrix2D(rotPoint, angle, 1.0)
    dimensions = (width, height)
    
    return cv.warpAffine(img, rotMat, dimensions)

rotated = rotate(img, 45)
plt.imshow(rotated)
```

```python
# Resize
resized = cv.resize(img, (500, 500), interpolation=cv.INTER_CUBIC)
plt.imshow(resized)
```

```python
#Flipping
flip = cv.flip(img, 0)
plt.imshow(flip)
```

```python
flip = cv.flip(img, 1)
plt.imshow(flip)
```

```python
flip = cv.flip(img, -1)
plt.imshow(flip)
```

```python
# Cropping
cropped = img[200:300, 300:400]
plt.imshow(cropped)
```
