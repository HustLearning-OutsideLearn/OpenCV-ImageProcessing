```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

resource_path = 'D:\GitCloneProject\OpenCV-ImageProcessing\Resources'
photo_path = resource_path + '\Photos'
video_path = resource_path + '\Videos'
face_path = resource_path + '\Faces'
```

```python
# Contour is the boudary of an object
# Contour is different from edge
```

```python
img = cv.imread(photo_path+'\cat.jpg')
plt.imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
canny = cv.Canny(img, 125, 175)
plt.imshow(canny)
```

```python
# Find Contours
contours, hierachies = cv.findContours(canny, cv.RETR_LIST, cv.CHAIN_APPROX_SIMPLE)

print(f'{len(contours)} contour(s) found!')
```

```python
cv.drawContours(img, contours, -1, (0,0,255), 1)
plt.imshow(img)
```

```python
ret, thresh = cv.threshold(img, 125, 255, cv.THRESH_BINARY)
plt.imshow(thresh)
```
