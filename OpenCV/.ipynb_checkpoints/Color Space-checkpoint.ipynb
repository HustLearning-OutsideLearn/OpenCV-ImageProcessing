```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

resource_path = 'D:\GitCloneProject\OpenCV-ImageProcessing\Resources'
photo_path = resource_path + '\Photos'
video_path = resource_path + '\Videos'
face_path = resource_path + '\Faces'
```

```python
img = cv.imread(photo_path+'\cat.jpg')
plt.imshow(img)
```

```python
# BGR --> GrayScale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
# BGR --> HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
plt.imshow(hsv)
```

```python
# BGR --> L*a*b
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
plt.imshow(lab)
```
