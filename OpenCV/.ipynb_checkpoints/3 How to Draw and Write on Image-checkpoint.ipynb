```python
blank = np.zeros((500, 500, 3)).astype("uint8")
blank.shape
```

```python
plt.imshow(blank)
```

```python
# Paint the image a certain color
blank[:] = 0,255,0
plt.imshow(blank)
```

```python
blank[:] = 255,0,0
plt.imshow(blank)
```

```python
blank = np.zeros((500, 500, 3)).astype("uint8")
blank[200:300, 300:400] = 255,0,0
plt.imshow(blank)
```

```python
# Draw a rectangle
cv.rectangle(blank, (0,0), (250,250), (0, 255, 0), thickness=2)
plt.imshow(blank)
```

```python
cv.rectangle(blank, (0,0), (250,500), (0, 255, 0), thickness=2)
plt.imshow(blank)
```

```python
cv.rectangle(blank, (0,0), (250,500), (0, 255, 0), thickness=cv.FILLED)
plt.imshow(blank)
```

```python
blank = np.zeros((500, 500, 3)).astype("uint8")
cv.rectangle(blank, (0,0), (250,500), (0, 255, 0), thickness=-1)
plt.imshow(blank)
```

```python
blank = np.zeros((500, 500, 3)).astype("uint8")
cv.rectangle(blank, (0, 0), (blank.shape[1]//2, blank.shape[0]//2), (0, 255, 0), thickness=-1)
plt.imshow(blank)
```

```python
# Draw Circle
cv.circle(blank, (250,250), 40, (255,0,0), thickness=3)
plt.imshow(blank)
```

```python
# Draw a line 
cv.line(blank, (0, 0), (250, 250), (255, 255, 255), thickness=3)
plt.imshow(blank)
```

```python
# Write text 
cv.putText(blank, "Hello", (255, 255), cv.FONT_HERSHEY_COMPLEX, 1.0, (255, 255, 255), 2)
plt.imshow(blank)
```

```python
# Write text 
cv.putText(blank, "Hello", (400, 400), cv.FONT_HERSHEY_COMPLEX, 1.0, (255, 255, 255), 2)
plt.imshow(blank)
```
