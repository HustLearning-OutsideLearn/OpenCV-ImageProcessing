```python
import cv2 as cv
import matplotlib.pyplot as plt

resource_path = 'D:\GitCloneProject\OpenCV-ImageProcessing\Resources'
photo_path = resource_path + '\Photos'
video_path = resource_path + '\Videos'
face_path = resource_path + '\Faces'
```

```python
img = cv.imread(photo_path + '\cat.jpg')
plt.imshow(img)
```

```python
# Rescale

def rescaleFrame(frame, scale=0.75):
    width = int(frame.shape[1]*scale)
    height = int(frame.shape[0]*scale)
    dimension = (width, height)
    return cv.resize(frame, dimension, interpolation = cv.INTER_AREA)

plt.imshow(rescaleFrame(img))
```

```python
# Change resolution
def changeRes(width, height):
    # Only work for live video
    img.set(3, width)
    img.set(4, height)

plt.imshow(img)
```
